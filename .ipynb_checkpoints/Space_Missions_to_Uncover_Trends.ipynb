**"Analyzing Global Space Missions to Uncover Trends, Performance, and Reliability"**

*Space exploration agencies and private companies across the globe launch hundreds of space missions each year. However, there is a lack of a centralized, data-driven perspective to assess mission success rates, trends over time, and operational efficiency. This leads to missed opportunities in understanding failures, planning better launch timelines, and allocating resources efficiently.*

**✅ Objective / Solution:**

*This project aims to analyze a historical dataset of global space missions using Python and data visualization libraries. The solution includes:*

*🧹 Data Cleaning: Removing unnecessary columns, fixing date formats, and standardizing status labels.*

*📊 Mission Status Analysis: Visualizing the success vs. failure rates to evaluate mission reliability.*

*🧩 Rocket Usage Trends: Showing the active vs. retired status of rockets using pie charts.*

*📈 Year-wise Mission Trends: Analyzing how the number of space missions evolved over time.*

*📍 Insightful Plots: Generating bar plots and line graphs to understand performance and patterns.*

*Through this analysis, space research organizations, analysts, and enthusiasts can gain valuable insights into space mission trends and improve future strategies.*



```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Set up visualization style
%matplotlib inline
plt.style.use('ggplot')
sns.set_palette("husl")
```

```python
# Load the dataset
df = pd.read_csv('Space_Corrected.csv')

# Display basic info
print("Dataset Shape:", df.shape)
display(df.head())
```

```python
# Data Cleaning
# Drop unnecessary columns
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')

# Convert Datum to datetime and extract year
df['Datum'] = pd.to_datetime(df['Datum'], utc=True, errors='coerce')
df['Year'] = df['Datum'].dt.year

# Clean Status Mission column
df['Status Mission'] = df['Status Mission'].str.strip()
```

```python
# 1. Basic Statistics and Distributions
plt.figure(figsize=(15, 10))

# Mission Status Distribution
plt.subplot(2, 2, 1)
status_counts = df['Status Mission'].value_counts()
status_counts.plot(kind='bar', color=['green', 'red', 'blue', 'orange'])
plt.title('Mission Status Distribution')
plt.ylabel('Count')
plt.xticks(rotation=45)
```

```python
# Rocket Status Distribution
plt.subplot(2, 2, 2)
rocket_status = df['Status Rocket'].value_counts()
rocket_status.plot(kind='pie', autopct='%1.1f%%', startangle=90)
plt.title('Rocket Status Distribution')
plt.ylabel('')
```

```python
# Missions Over Time
plt.subplot(2, 2, 3)
missions_per_year = df['Year'].value_counts().sort_index()
missions_per_year.plot(kind='line', marker='o')
plt.title('Number of Missions Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Missions')
```

```python
# Top Companies by Number of Missions
plt.subplot(2, 2, 4)
top_companies = df['Company Name'].value_counts().nlargest(10)
top_companies.plot(kind='barh')
plt.title('Top 10 Companies by Number of Missions')
plt.xlabel('Number of Missions')

plt.tight_layout()
plt.show()
```

```python
# 2. Interactive Visualization with Plotly
# Missions by Company Over Time
missions_by_company = df.groupby(['Year', 'Company Name']).size().reset_index(name='Count')
top_10_companies = df['Company Name'].value_counts().nlargest(10).index.tolist()
filtered_missions = missions_by_company[missions_by_company['Company Name'].isin(top_10_companies)]

fig = px.line(filtered_missions, x='Year', y='Count', color='Company Name',
              title='Missions Over Time by Top 10 Companies',
              labels={'Count': 'Number of Missions'},
              template='plotly_white')
fig.update_layout(hovermode='x unified')
fig.show()
```

```python
# 3. Geographical Distribution
location_counts = df['Location'].value_counts().reset_index()
location_counts.columns = ['Location', 'Count']
```

```python
# Extract country from location (simple extraction)
location_counts['Country'] = location_counts['Location'].str.split(',').str[-1].str.strip()

country_counts = location_counts.groupby('Country')['Count'].sum().reset_index().sort_values('Count', ascending=False)

fig = px.choropleth(country_counts,
                    locations='Country',
                    locationmode='country names',
                    color='Count',
                    hover_name='Country',
                    color_continuous_scale=px.colors.sequential.Plasma,
                    title='Space Missions by Country')
fig.show()
```

```python
# 4. Mission Success Rate by Company
success_rates = df.groupby('Company Name')['Status Mission'].apply(
    lambda x: (x == 'Success').mean() * 100).reset_index(name='Success Rate')

top_20_companies = df['Company Name'].value_counts().nlargest(20).index
success_rates_top = success_rates[success_rates['Company Name'].isin(top_20_companies)].sort_values('Success Rate', ascending=False)

plt.figure(figsize=(12, 8))
sns.barplot(x='Success Rate', y='Company Name', data=success_rates_top, palette='viridis')
plt.title('Success Rate of Top 20 Most Active Companies')
plt.xlabel('Success Rate (%)')
plt.ylabel('Company Name')
plt.tight_layout()
plt.show()
```

```python
# 5. Rocket Status Analysis
rocket_status_by_company = df.groupby(['Company Name', 'Status Rocket']).size().unstack().fillna(0)
rocket_status_by_company['Total'] = rocket_status_by_company.sum(axis=1)
top_rocket_companies = rocket_status_by_company.nlargest(10, 'Total')

plt.figure(figsize=(12, 8))
top_rocket_companies.drop('Total', axis=1).plot(kind='barh', stacked=True, figsize=(12, 8))
plt.title('Rocket Status for Top 10 Companies by Rocket Usage')
plt.xlabel('Number of Rockets')
plt.ylabel('Company Name')
plt.legend(title='Rocket Status')
plt.tight_layout()
plt.show()
```

```python
# 6. Time Series Analysis with Success/Failure
df['Mission Success'] = df['Status Mission'] == 'Success'
time_series = df.groupby('Year')['Mission Success'].agg(['mean', 'count']).reset_index()
time_series['Success Rate'] = time_series['mean'] * 100
time_series['Total Missions'] = time_series['count']

fig = make_subplots(specs=[[{"secondary_y": True}]])

fig.add_trace(
    go.Scatter(x=time_series['Year'], y=time_series['Success Rate'], name="Success Rate"),
    secondary_y=False,
)

fig.add_trace(
    go.Bar(x=time_series['Year'], y=time_series['Total Missions'], name="Total Missions", opacity=0.3),
    secondary_y=True,
)

fig.update_layout(
    title_text="Mission Success Rate and Total Missions Over Time",
    template='plotly_white'
)

fig.update_xaxes(title_text="Year")
fig.update_yaxes(title_text="Success Rate (%)", secondary_y=False)
fig.update_yaxes(title_text="Number of Missions", secondary_y=True)

fig.show()
```

```python
# Import necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re  # Needed for failure reasons analysis
from datetime import datetime
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Load the dataset
df = pd.read_csv('Space_Corrected.csv')
```

```python
# Data Cleaning
# Drop unnecessary columns
df = df.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1, errors='ignore')

# Convert Datum to datetime and extract year
df['Datum'] = pd.to_datetime(df['Datum'], utc=True, errors='coerce')
df['Year'] = df['Datum'].dt.year
```

```python
# Clean Status Mission column
df['Status Mission'] = df['Status Mission'].str.strip()
```

```python
# Comparative Analysis of Successful vs Failed Missions

# Create a new column for simplified mission status
df['Mission_Status'] = df['Status Mission'].apply(lambda x: 'Success' if x == 'Success' else 'Failure')
```

```python
# 1. Success vs Failure Over Time
plt.figure(figsize=(15, 6))
status_over_time = df.groupby(['Year', 'Mission_Status']).size().unstack()
status_over_time.plot(kind='line', marker='o', ax=plt.gca())
plt.title('Success vs Failure Over Time')
plt.ylabel('Number of Missions')
plt.xlabel('Year')
plt.grid(True)
plt.legend(title='Mission Status')
plt.show()
```

```python
# 2. Success vs Failure by Company (Top 20)
top_companies = df['Company Name'].value_counts().nlargest(20).index
company_status = df[df['Company Name'].isin(top_companies)].groupby(
    ['Company Name', 'Mission_Status']).size().unstack().fillna(0)
company_status['Total'] = company_status.sum(axis=1)
company_status = company_status.sort_values('Total', ascending=False)

plt.figure(figsize=(15, 8))
company_status[['Success', 'Failure']].plot(kind='barh', stacked=True, ax=plt.gca())
plt.title('Success vs Failure for Top 20 Companies')
plt.xlabel('Number of Missions')
plt.ylabel('Company Name')
plt.legend(title='Mission Status')
plt.tight_layout()
plt.show()
```

```python
# 3. Success Rate by Decade
df['Decade'] = (df['Year'] // 10) * 10
success_by_decade = df.groupby('Decade')['Mission_Status'].apply(
    lambda x: (x == 'Success').mean() * 100).reset_index(name='Success Rate')

plt.figure(figsize=(10, 6))
sns.lineplot(data=success_by_decade, x='Decade', y='Success Rate', marker='o')
plt.title('Success Rate by Decade')
plt.ylabel('Success Rate (%)')
plt.xlabel('Decade')
plt.ylim(0, 100)
plt.grid(True)
plt.show()
```

```python
# 4. Interactive Sunburst Chart (Company > Status)
status_counts = df.groupby(['Company Name', 'Mission_Status']).size().reset_index(name='Count')
top_30_companies = df['Company Name'].value_counts().nlargest(30).index
status_counts = status_counts[status_counts['Company Name'].isin(top_30_companies)]

fig = px.sunburst(status_counts, path=['Company Name', 'Mission_Status'], values='Count',
                  color='Mission_Status', color_discrete_map={'Success':'green','Failure':'red'},
                  title='Mission Status Distribution by Company')
fig.update_traces(textinfo="label+percent parent")
fig.show()
```

```python
# 5. Failure Reasons Analysis (when available)
if 'Detail' in df.columns:
    # Extract failure reasons from details
    failure_df = df[df['Mission_Status'] == 'Failure']

    # Modified extraction to handle cases with no matches
    failure_reasons = failure_df['Detail'].str.extract(r'(fail|explod|destroy|abort|lost|crash)', flags=re.IGNORECASE)

    # Drop rows with no matches (NaN values) before accessing column
    failure_reasons = failure_reasons.dropna()

    # Only proceed if we have matches
    if not failure_reasons.empty:
        # Now we can safely access the column
        counts = failure_reasons[0].str.lower().value_counts().head(10)

        plt.figure(figsize=(12, 6))
        counts.plot(kind='barh', color='red')
        plt.title('Top 10 Failure Reasons Mentioned in Mission Details')
        plt.xlabel('Count')
        plt.ylabel('Failure Reason')
        plt.show()
    else:
        print("No failure reasons found in the data.")
```

```python
# 6. Side-by-Side Comparison of Rocket Status vs Mission Status
cross_tab = pd.crosstab(df['Status Rocket'], df['Mission_Status'], normalize='index') * 100

plt.figure(figsize=(10, 6))
sns.heatmap(cross_tab, annot=True, fmt='.1f', cmap='RdYlGn', vmin=0, vmax=100)
plt.title('Rocket Status vs Mission Success (%)')
plt.ylabel('Rocket Status')
plt.xlabel('Mission Status')
plt.tight_layout()
plt.show()
```
